- **Temporal Coding**: The time the neuron fires is the information that is encoded
- **Rate Coding**: Amount of neuron firing at a particular time
- **Latency**: The initial lag of having firings in correlation to stimulus
- **Synchronous**: Multiple neurons firing at once creating coincidence spikes 
- **Tuning Codes**: Characterizing neuron patterns from stimulus
- **Population Codes**: Generated by combining information from multiple neurons

***

Temporal Coding deals with timing while rate coding focuses on the amount of neuronal firing. Latency is the initial lag while synchornous is the opposite of haveing an amplified chart because of multiple neurons firing at the same time. Population codes include the noise that tuning codes do not have.
